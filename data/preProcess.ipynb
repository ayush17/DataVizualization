```python
#Group the data by 'attacktype1_txt'
attack_type_data = df.groupby(
	'attacktype1_txt').size().reset_index(name='value')
attack_type_data = attack_type_data.rename(columns={'attacktype1_txt': 'name'})

attack_type_sunburst = [{
	'name': 'Attack Type',
	'children': attack_type_data.to_dict('records')
}]

#Saving the data to a JSON file
with open('attack_type_sunburst.json', 'w') as f:
	json.dump(attack_type_sunburst, f, indent=4)

df['continent_txt'] = df['country_txt'].apply(
	lambda x: continents.get(x, 'Unknown'))

suicide_attack_data = df[df['suicide'] == 1].groupby(
	['continent_txt', 'country_txt', 'region_txt']).size().reset_index(name='value')

suicide_attack_sunburst = [{
	'name': 'Suicide Attacks',
	'children': [
		{
			'name': continent,
			'children': [
				{
					'name': region,
					'children': [
						{
							'name': country,
							'value': value
						} for country, value in suicide_attack_data[(suicide_attack_data['continent_txt'] == continent) & (suicide_attack_data['region_txt'] == region)][['country_txt', 'value']].values
					]
				} for region in suicide_attack_data[suicide_attack_data['continent_txt'] == continent]['region_txt'].unique()
			]
		} for continent in suicide_attack_data['continent_txt'].unique()
	]
}]
#Saving the data to a JSON file
with open('suicide_attack_sunburst.json', 'w') as f:
	json.dump(suicide_attack_sunburst, f, indent=4)
```
