```python
import pandas as pd

df=pd.read_csv("https://raw.githubusercontent.com/polakowo/socialdata2017/master/project/jupyter/data/globalterrorism_cleaned.csv")
df
```

```python
# Group by country code and sum the killed column
killed_by_country = df.groupby(['countrycode', 'country','year'])['nkilled'].sum().reset_index()


killed_by_country=killed_by_country.sort_values(by='year',ascending=True)
killed_by_country.to_csv("killed_by_country.csv",index=False)
```
